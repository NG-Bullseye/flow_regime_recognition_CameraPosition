```python
from data_import_and_preprocessing.dataset_generator import DatasetGenerator
import cv2
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
class LabelDatasetGenerator(DatasetGenerator):
    def read_image(self, file):
        return None

    def read_label(self, file):
        with open(file) as f:
            json_content = json.load(f)
            label_int = json_content['labels']['flow_regime']['value']
            return label_int
```

```python
data_dir = '/mnt/0A60B2CB60B2BD2F/Datasets/bioreactor_flow_regimes/02_data'
label_data_generator = LabelDatasetGenerator(data_dir, img_shape=(192, 108, 1), no_classes=3)
labels = []
for _, label in label_data_generator.get_data_point_generator():
    labels.append(label)
```

```python
plt.hist(labels, bins=3)
```

```python
class SampleDatasetGenerator(DatasetGenerator):
    def read_label(self, file):
        with open(file) as f:
            json_content = json.load(f)
            label_int = json_content['labels']['flow_regime']['value']
            return label_int

data_dir = '/mnt/0A60B2CB60B2BD2F/Datasets/bioreactor_flow_regimes/02_data'
sample_data_generator = SampleDatasetGenerator(data_dir, img_shape=(192, 108, 1), no_classes=3, no_points=1000)

fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(16, 16))
axs = axs.flatten()

labels = [0, 0, 0, 1, 1, 1, 2, 2, 2]
images = []

no = 0
for point_image, point_label in sample_data_generator.get_data_point_generator():
    #print(f'{no}: label is {point_label}')
    if point_label is labels[no]:
        images.append(point_image)
        no += 1
        if no == 9:
            break

for label, image, ax in zip(labels, images, axs):
    ax.imshow(image, cmap='gray', vmin=0, vmax=1)
    font_axis_publish = {
          'color':  'black',
          'weight': 'bold',
          'size': 12,
    }
    ax.axis('off')
    ax.set_title(str(label), fontdict=font_axis_publish)

plt.tight_layout()
```
